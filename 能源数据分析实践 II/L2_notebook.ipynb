## 标题与文本格式化

### 标题示例

- `## 二级标题`

## 数学公式
Jupyter Notebook 支持 LaTeX 数学公式，既可以嵌入到行内，也可以单独展示。

### 行内公式
使用 `$公式内容$` 表示，例如：$y = mx + b$

### 独占一行公式
使用 `$$公式内容$$` 表示：$$y = x^2 + 3x + 5$$

### 复杂公式示例
$$\frac{\sqrt{x^2 + y^2}}{z}$$
$$\sum_{i=1}^n x_i$$



## 数据可视化
Jupyter Notebook 提供了强大的可视化能力，以下是一个简单的示例:


````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
plt.title('Sine Function')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
````
